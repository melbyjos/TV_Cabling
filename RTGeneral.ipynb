```python
import numpy as np
import scipy as sp
import time
import math

import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression


np.set_printoptions(precision=2)
```

```python
def getkjp(r,q,l):
    m = (r-1)//2
    ip = q*l+1
    k = ip//r
    j = ip%r
    if j < m+1:
        return j
    else:
        return r-j
def getkjm(r,q,l):
    m = (r-1)//2
    ip = q*l-1
    k = ip//r
    j = ip%r
    if j < m+1:
        return j
    else:
        return r-j

def RT1s(r,q):
    t0 = time.time()
    m = (r-1)//2
    #print(m)
    mat = np.zeros(shape=(m,m))
    for l in range(0,m):
        col = np.zeros(shape=(m,))
        jp = getkjp(r,q,l)
        jm = getkjm(r,q,l)
        if jp == jm == 1:
            col[0] = 1
        if jm == 0 and jp == 2:    
            col[jp-1] = 1
        if jp == 0 and jm == 2:    
            col[jm-1] = -1
        if jp + jm > 2:
            col[jp-1] = 1
            col[jm-1] =- 1
        mat[:,l] = col
    return mat

def qp6Bound(q):
    nums = []
    if (q%2) == 0:
        for r in range(3,q+3+1,2):
            if math.gcd(r,q) == 1:
                nums.append(r)
    else:
        for r in range(3,q+6+1,2):
            if math.gcd(r,q) == 1:
                nums.append(r)
    return nums

def qm3Bound(q):
    nums = []
    if q%2 == 0:
        for r in range(3,q,2):
            if math.gcd(r,q) == 1:
                nums.append(r)
    return nums
```
